# PLOTTING

```python
f, ax = plt.subplots(figsize=(16,12))
sns.heatmap(data.corr(), annot=True)
```

```python
fig, ax = plt.subplots(figsize=(14,10))
ax.set_title('DISTR OF RATINGS', fontsize=16)
ax.set_xlabel("RATING", fontsize=16)
ax.set_ylabel('DISTR', fontsize=16)

data['ratings'].value_counts().plot(ax=ax, kind='bar', color='red')
```
