```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
from scipy.stats import pearsonr
```

```python
mcu_movie_data = pd.read_csv('./input/movie_info.csv')
mcu_movie_data.head()
```

```python
mcu_movie_data.sort_values(by="production_budget_in_million_(USD)", ascending=False).head(2)
```

```python
round(np.mean(mcu_movie_data['production_budget_in_million_(USD)']), 2)
```

```python
avarage_ratings = []

for i in range(len(mcu_movie_data)):
    rat_list = []
    rat_list.append(mcu_movie_data['tomatometer'][i])
    rat_list.append(mcu_movie_data['tomato_audience_score'][i])
    rat_list.append(mcu_movie_data['imdb'][i])
    rat_list.append(mcu_movie_data['metascore'][i])
    rat_list.append(mcu_movie_data['meta_user_score'][i])
    
    avarage_ratings.append(round(np.mean(rat_list), 1))
    
mcu_movie_data['avarage_ratings'] = avarage_ratings

mcu_movie_data.head()
```

```python
mcu_movie_data.sort_values(by='avarage_ratings', ascending=False).head()
```

```python
mcu_movie_data.sort_values(by='avarage_ratings').head()
```

```python
round(np.mean(mcu_movie_data['avarage_ratings']), 1)
```

```python
mcu_movie_data.sort_values(by="worldwide_collection_in_million_(USD)",ascending=False).head()
```

```python
mcu_movie_data.sort_values(by="worldwide_collection_in_million_(USD)").head()
```

```python
earnings_df = {'movie_title': mcu_movie_data['movie_title'], 'worldwide_collection': mcu_movie_data['worldwide_collection_in_million_(USD)']}

sorted_earnings_df = pd.DataFrame(data=earnings_df)
sorted_earnings_df = sorted_earnings_df.sort_values(by='worldwide_collection', ascending=False)
plt.figure(figsize=(15, 20))

graph_1 = sns.barplot(data=sorted_earnings_df, x='worldwide_collection', y='movie_title', orient='h')
graph_1.set_xlabel('Collection in million (USD)', fontsize=20)
graph_1.set_ylabel('Movie Title', fontsize=20)
graph_1.set_title('MCU box collection', fontsize=20)

graph_1.bar_label(graph_1.containers[0])
plt.show()
```

```python
r_value, p_value = pearsonr(mcu_movie_data['production_budget_in_million_(USD)'], mcu_movie_data['worldwide_collection_in_million_(USD)'])
```

```python
r_value = round(r_value, 4)
r_value
```

```python
fig, graph_2 = plt.subplots(figsize=(10, 6))

sns.regplot(data=mcu_movie_data, x='production_budget_in_million_(USD)', y='worldwide_collection_in_million_(USD)')

graph_2.set_xlim(0, )
graph_2.set_ylim(0, )

graph_2.set_xlabel('Production Budget (m USD)', fontsize=15)
graph_2.set_ylabel('Worldwide Collection (m USD)', fontsize=15)

graph_2.text(x=0.5, y=1.1, s='Budget vs Box Office Collection', fontsize=20, weight='bold', ha='center', va='bottom', transform=graph_2.transAxes)
graph_2.text(x=0.5, y=1.05, s= f'Pearson correlation (r) = {r_value}', fontsize=15, alpha=0.75, ha='center', va='bottom', transform=graph_2.transAxes)
```
