```python
# This notebook is designed to solve the classic problem of classifying iris flowers, but using DABL
# Dabl is cool Data Analysis Baseline Library
# Now I'll show you why
```

```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import dabl
```

```python
# I chose this dataset because of its simplicity and clarity. In general, this is just an example.

iris_data = pd.read_csv('./input/Iris.csv')
iris_data.head()
```

```python
# We split our data into 'X' and 'y'. Standard procedure, nothing special.

X = iris_data.drop('Species', axis=1)
y = iris_data['Species']
```

```python
# Now it's time to double. Using the 'double.plot' method, you can get rid of tedious plotting. This guy will do everything himself.
dabl.plot(X, y)
```

```python
# We looked at the charts. Now let's build the model. 'train_test_split' Yes, nothing special
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# And again on stage 'dabl'. Simple_Classifier will check all popular algorithms for you and select the optimal one. This is unrealistically cool for me.
dabl_simplifier = dabl.SimpleClassifier()
dabl_simplifier.fit(X_train, y_train)
```

```python
# Ready. The accuracy is about 98%. not bad. And the coolest thing is that it's all in less than 5 minutes.

dabl_simplifier.score(X_test, y_test)
```

```python
# I use sklearn.metrics to check the accuracy of the model.
from sklearn.metrics import accuracy_score
```

```python
# find predicted values

y_pred = dabl_simplifier.predict(X)
y_pred = pd.Series(y_pred)
```

```python
accuracy_score(y, y_pred)
```
